```python
from ultralytics import YOLO
from PIL import Image
```

```python
import os

def list_files_in_directory(directory):

    file_paths = []
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            if os.path.isfile(file_path):
                file_paths.append(file_path)
    return file_paths
```

```python
test_images = list_files_in_directory("test_images")
```

```python
model_cls = YOLO("custom_models/best_30e_cls.pt")
model_seg = YOLO("custom_models/best_30e_seg.pt")
model_det = YOLO("custom_models/last_30e_det.pt")
```

```python
index = 1
for x in test_images[:1]:
  results = model_det("mit2.jpg")
  for result in results:
      original_image = result.orig_img
      boxes = result.boxes.xyxy

      index2 = 1

      for bb in boxes:
        x, y, w, h = bb
        x, y, w, h = int(x), int(y), int(w), int(h)
        # print(x,y,w,h)
        det_image = original_image[y:h , x:w]

        image = Image.fromarray(det_image)

        det_path = f'det_results/det{index}.{index2}.jpg'

        image.save(det_path)

        seg_results = model_seg(det_path)
        for seg_result in seg_results:
          # seg_result.show()  
          seg_result.save(filename=f'seg_results/seg{index}.{index2}.jpg')  

        cls_results = model_cls(det_path)
        for cls_result in cls_results:
          # cls_result.show()  
          cls_result.save(filename=f'cls_results/cls{index}.{index2}.jpg')  
        
        index2 += 1

      probs = result.probs  
      # result.show()  
      result.save(filename=f"det_results/det{index}.jpg")  

  index += 1
```
